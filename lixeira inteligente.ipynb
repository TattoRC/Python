```python
import random

# Esse paragrafo tem a função de calcular a quantidades de brindes por quantidades de lixo
def calcular_brindes(quantidade_lixo):
    return quantidade_lixo // 10

# Função principal do sistema de lixeira
def sistema_lixeira():
    try:
        # Base de dados 
        usuarios = []
        
        # Menu de base, incluindo a criação de usuarios ou ja fazer seu login para ver o menu principal 
        while True:
            print("\nMenu Principal:")
            print("1. Registrar novo usuário")
            print("2. Entrar")
            print("3. Sair")

            opcao_principal = input("Escolha uma opção: ")

            if opcao_principal == '1':
                usuario_id = input("Digite seu novo ID: ")
                if any(usuario['id'] == usuario_id for usuario in usuarios):
                    print("ID já existente. Escolha outro ID.")
                    continue
                
                senha = input("Digite uma senha de 4 dígitos numéricos: ")
                if not senha.isdigit() or len(senha) != 4:
                    print("Senha inválida. A senha deve conter exatamente 4 dígitos numéricos.")
                    continue
                
                usuarios.append({'id': usuario_id, 'senha': senha, 'lixeira': []})
                print("Usuário registrado com sucesso!")
            
            elif opcao_principal == '2':
                usuario_id = input("Digite seu ID: ")
                usuario = next((usuario for usuario in usuarios if usuario['id'] == usuario_id), None)
                
                if usuario is None:
                    print("Usuário não encontrado.")
                    continue
                
                senha = input("Digite sua senha: ")
                if senha != usuario['senha']:
                    print("Senha incorreta.")
                    continue

                # Aqui se inicia o menu principal, onde ira ter acesso as seguintes opções: 
                # adicionar lixo, Ver a quantidade de lixo e o que são os lixos, 3 ver a quantidade de brindes que você possue pela quantidade de lixos depositados.
                
                print(f"Bem-vindo, {usuario_id}!")
                
                while True:
                    print("\nMenu do Usuário:")
                    print("1. Adicionar lixo")
                    print("2. Ver lixeira")
                    print("3. Ver quantidade de brindes")
                    print("4. Sair")

                    opcao_usuario = input("Escolha uma opção: ")

                    if opcao_usuario == '1':
                        lixo = input("Digite o que deseja jogar na lixeira: ")
                        usuario['lixeira'].append(lixo)
                        print("Lixo adicionado com sucesso!")
                    elif opcao_usuario == '2':
                        if not usuario['lixeira']:
                            print("A lixeira está vazia.")
                        else:
                            print("Conteúdo da lixeira:")
                            for i, item in enumerate(usuario['lixeira'], start=1):
                                print(f"{i}. {item}")
                    elif opcao_usuario == '3':
                        quantidade_brindes = calcular_brindes(len(usuario['lixeira']))
                        print(f"Você tem {quantidade_brindes} brindes.")
                    elif opcao_usuario == '4':
                        print("Saindo do sistema...")
                        break
                    else:
                        print("Opção inválida. Tente novamente.")
            
            elif opcao_principal == '3':
                print("Saindo do sistema...")
                break
            else:
                print("Opção inválida. Tente novamente.")

    except Exception as e:
        print(f"Ocorreu um erro: {e}")

if __name__ == "__main__":
    sistema_lixeira()
```
